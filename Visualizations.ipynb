Importing the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the dataset

```python
#Data which shows the return of the vanilla model at a risk of 15%
Vanilla = pd.read_excel("Vanilla_Results.xlsx")

#Data which shows the return of the quantum model at a risk of 15%
Quantum = pd.read_excel("Quantum_Results.xlsx")
```

Plotting the return comparisons

```python
# Convert the return of the quantum results to a percentage
Quantum['return'] *= 100
```

```python
# Combine data for plotting
data = pd.DataFrame({
    'Type': ['Vanilla', 'Quantum'],
    'Return': [Vanilla['return'].iloc[0], Quantum['return'].iloc[0]]
})
```

```python
# Plotting
plt.figure(figsize=(8, 4))
bars = plt.bar(data['Type'], data['Return'], color=['blue', 'green'])

# Adding value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}')

plt.ylabel('Return')
plt.title('Return Values Comparison for a risk of 1%')
plt.show()
```
