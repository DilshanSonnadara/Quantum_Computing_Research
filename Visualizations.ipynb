Importing the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Importing the dataset

```python
#Asset Allocation Data
Vanilla = pd.read_excel("Vanilla_Results.xlsx")

#Results Data
Quantum = pd.read_excel("Quantum_Results.xlsx")
```

Plotting the return comparisons

```python
# Modify the Quantum return value
Quantum['return'] *= 100
```

```python
# Combine data for plotting
data = pd.DataFrame({
    'Type': ['Vanilla', 'Quantum'],
    'Return': [Vanilla['return'].iloc[0], Quantum['return'].iloc[0]]
})

# Plotting
plt.figure(figsize=(8, 4))
plt.bar(data['Type'], data['Return'], color=['blue', 'green'])


bars = plt.bar(data['Type'], data['Return'], color=['blue', 'green'])

# Adding value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')


plt.ylabel('Return')
plt.title('Return Values Comparison for a risk of 1%')
plt.show()
```
