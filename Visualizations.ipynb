Importing the Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

Importing the dataset

```python
#Asset Allocation Data
Assets = pd.read_excel("Assets1.xlsx")

#Results Data
Results = pd.read_excel("Final_Results.xlsx")

#Profits Data
Profits = pd.read_excel("Cleaned_Data_Index_1.xlsx")
```

Plotting the asset allocation chart

```python
# Create the pie chart
fig = px.pie(Assets, names='Asset', values='Value', title='Number of Units of Assets to Purchase')
fig.update_traces(textinfo='value')

# Show the plot
fig.show()
```

Gettings the correlation heatmap

```python
# Columns to include in the correlation matrix
columns_to_include = Assets["Asset"].tolist() 

# Calculate the correlation matrix for the specified columns
corr_matrix = Profits [columns_to_include].corr()

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

Efficient Frontier

```python
# Create the Efficient Frontier line chart
fig = px.line(Results, x='R', y='total_profit', title='Efficient Frontier',
              labels={'R': 'Maximum Allowable Risk', 'total': 'Total Return'},
              markers=True)

# Show the plot
fig.show()
```

Time Taken

```python
# Create the Time line chart
fig = px.line(Results, x='R', y='time_taken', title='Time Taken to Calculate Allocations',
              labels={'R': 'Maximum Allowable Risk', 'time_taken': 'Total Time'},
              markers=True)

# Show the plot
fig.show()
```
